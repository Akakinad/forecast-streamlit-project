```python
import gc

# Initialize a counter to accumulate the results
store_product_count = {}

# Load the dataset in chunks
chunk_size = 100000  # Adjust the chunk size if necessary
chunks = pd.read_csv('/content/drive/MyDrive/kaggle_data/train.csv', chunksize=chunk_size)

# Process each chunk individually
for chunk in chunks:
    # You can apply filtering or processing directly to the chunk
    for store in chunk['store_nbr'].unique():
        store_data = chunk[chunk['store_nbr'] == store]
        num_products_sold = store_data['item_nbr'].nunique()

        # Store the count in a dictionary (or perform any other aggregation)
        store_product_count[store] = store_product_count.get(store, 0) + num_products_sold

    # Delete the chunk after processing to free up memory
    del chunk
    gc.collect()  # Force garbage collection to clear up unused memory

# After processing, you can analyze the `store_product_count` dictionary
print(store_product_count)
```

```python
# Extract only stores in Guayas
df_stores_guayas = df_stores[df_stores['state'] == 'Guayas']

# Calculate average products sold per store in Guayas
guayas_store_product_count = {store: count for store, count in store_product_count.items() if store in df_stores_guayas['store_nbr'].values}

# Calculate the average number of products sold per store
average_products_sold = sum(guayas_store_product_count.values()) / len(guayas_store_product_count)

print(f"Average products sold per store in Guayas: {average_products_sold}")
```

```python
import pandas as pd
import gc

# Initialize the counter for new items
new_items = []

# Load the dataset in chunks
chunk_size = 100000  # Adjust the chunk size as necessary
chunks = pd.read_csv('/content/drive/MyDrive/kaggle_data/train.csv', chunksize=chunk_size)

# Process each chunk individually
for chunk in chunks:
    # Convert the 'date' column to datetime format
    chunk['date'] = pd.to_datetime(chunk['date'])

    # Filter the chunk for stores in Guayas
    df_guayas_chunk = chunk[chunk['store_nbr'].isin(df_stores_guayas['store_nbr'].values)]

    # Sort by store, item, and date to track the first appearance of each item
    new_items_chunk = df_guayas_chunk.sort_values(by=['store_nbr', 'item_nbr', 'date']).drop_duplicates(subset=['store_nbr', 'item_nbr'], keep='first')

    # Append the result to the new_items list
    new_items.append(new_items_chunk)

    # Free memory after processing the chunk
    del chunk
    gc.collect()

# Concatenate the chunks together into one DataFrame
new_items_df = pd.concat(new_items, ignore_index=True)

# Now count the number of unique new items introduced in each store
new_items_count = new_items_df.groupby('store_nbr')['item_nbr'].nunique()

# Display the results
print("Number of new items introduced in each store during January-March 2014:")
print(new_items_count)
```

```python
# Define the top-selling items to filter
top_item_nbrs = top_selling_items_df['item_nbr'].values  # Top-selling item numbers from previous calculation

# Initialize an empty dictionary to store total sales for top items
top_items_sales = {}

# Process the data in chunks
chunk_size = 100000  # Size of each chunk (adjust if necessary)
chunks = pd.read_csv('/content/drive/MyDrive/kaggle_data/train.csv', chunksize=chunk_size)

# Loop through each chunk
for chunk in chunks:
    # Filter the chunk for the top-selling items
    chunk_top_items = chunk[chunk['item_nbr'].isin(top_item_nbrs)]

    # Calculate the total sales (unit_sales) per item (group by item_nbr)
    sales_per_item = chunk_top_items.groupby('item_nbr')['unit_sales'].sum()

    # Accumulate the sales per item
    for item_nbr, sales in sales_per_item.items():
        if item_nbr in top_items_sales:
            top_items_sales[item_nbr] += sales
        else:
            top_items_sales[item_nbr] = sales

# Convert the dictionary to a DataFrame
top_selling_items_df = pd.DataFrame(list(top_items_sales.items()), columns=['item_nbr', 'unit_sales'])

# Sort the items by total unit_sales and get the top N items
top_n = 10  # For example, top 10 items
top_selling_items_df = top_selling_items_df.sort_values(by='unit_sales', ascending=False).head(top_n)

# Show the top-selling items
print(top_selling_items_df)
```

#Filter the transactions data for the top-selling items

```python
# Initialize a dictionary to accumulate time-series data for top-selling items
time_series_data = {}

# Process the data in chunks
chunk_size = 100000  # Adjust the chunk size as necessary
chunks = pd.read_csv('/content/drive/MyDrive/kaggle_data/train.csv', chunksize=chunk_size)

# Loop through each chunk
for chunk in chunks:
    # Filter the chunk for the top-selling items
    chunk_top_items = chunk[chunk['item_nbr'].isin(top_selling_items_df['item_nbr'].values)]

    # Aggregate sales data (sum unit_sales per date and item_nbr)
    time_series_chunk = chunk_top_items.groupby(['date', 'item_nbr'])['unit_sales'].sum().reset_index()

    # Accumulate the time-series data for top items
    for _, row in time_series_chunk.iterrows():
        item_nbr = row['item_nbr']
        date = row['date']
        unit_sales = row['unit_sales']

        # Initialize the dictionary for each item if not already done
        if item_nbr not in time_series_data:
            time_series_data[item_nbr] = []

        # Append the sales data to the item list
        time_series_data[item_nbr].append((date, unit_sales))

# Convert the accumulated data into a DataFrame for each top-selling item
time_series_df_list = []
for item_nbr, data in time_series_data.items():
    temp_df = pd.DataFrame(data, columns=['date', 'unit_sales'])
    temp_df['item_nbr'] = item_nbr
    time_series_df_list.append(temp_df)

# Concatenate the individual DataFrames for each item
df_time_series = pd.concat(time_series_df_list, ignore_index=True)

# Convert the 'date' column to datetime format
df_time_series['date'] = pd.to_datetime(df_time_series['date'])

# Check the first few rows to confirm
df_time_series.head()
```

#Key points of the steps done above:

We filter each chunk for only the top-selling items.

We aggregate unit sales for each combination of date and item number.

The time-series data is accumulated in chunks to avoid memory overload.

Finally, we combine the results for all chunks into one DataFrame (df_time_series), where each row represents sales data for a specific item on a specific date.

#Decompose the Time Series

We can decompose the time series data to better understand the trend, seasonality, and residual (noise) components of the sales data for the top-selling items.


The goal of decomposition is to break the time series data into:


Trend: The long-term movement or direction in the data.

Seasonality: The repeating short-term patterns or cycles (e.g., monthly or yearly).

Residual: The remaining noise or irregular fluctuations after removing the trend and seasonality.

We need the sales data to be in a DateTime index format, which will make decomposition easier. So, we will first ensure the month column is converted to a datetime format.



```python
# Convert 'month' column from PeriodDtype to datetime
df_monthly_sales['month'] = df_monthly_sales['month'].dt.to_timestamp()

# Set the 'month' column as the index
df_monthly_sales.set_index('month', inplace=True)
```

```python
from statsmodels.tsa.seasonal import STL

# Select sales data for one item (e.g., item 215352)
item_data = df_monthly_sales[df_monthly_sales['item_nbr'] == 215352]['unit_sales']

# Decompose the time series (using STL decomposition)
decomposition = STL(item_data, period=12).fit()  # 12 months period for yearly seasonality

# Extract trend, seasonal, and residual components
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot the decomposition components
plt.figure(figsize=(12, 10))

plt.subplot(4, 1, 1)
plt.plot(item_data, label='Original')
plt.title(f'Original Sales Data for Item {215352}')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(trend, label='Trend', color='orange')
plt.title('Trend Component')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(seasonal, label='Seasonality', color='green')
plt.title('Seasonality Component')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(residual, label='Residual', color='red')
plt.title('Residual Component')
plt.legend()

plt.tight_layout()
plt.show()
```

# Interpretation
The plot will show:

The original sales data for item 215352.

The trend component shows the long-term changes in sales (whether they are generally increasing or decreasing).

The seasonality component highlights any regular monthly or yearly patterns.

The residual component captures the random noise or irregular fluctuations after removing trend and seasonality.

```python
# Pivot the data to have 'item_nbr' as columns and 'month' as index
df_pivot = df_monthly_sales.pivot_table(index='month', columns='item_nbr', values='unit_sales', aggfunc='sum')

# Check the shape of the pivoted DataFrame
print(df_pivot.head())
```

#Apply Time-Series Decomposition for Each Item

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Initialize the plot
plt.figure(figsize=(12, 8))

# Loop through each item and perform decomposition
for item_nbr in df_pivot.columns:
    # Get the time series for the current item
    item_sales = df_pivot[item_nbr].dropna()  # Drop any missing values

    # Decompose the time series
    decomposition = seasonal_decompose(item_sales, model='multiplicative', period=12)

    # Plot the decomposition for each item
    plt.subplot(len(df_pivot.columns), 1, list(df_pivot.columns).index(item_nbr) + 1)
    decomposition.plot()
    plt.title(f'Decomposition for Item {item_nbr}')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
